```python
import pyautogui as pa
import pyperclip as pc
import time
```

```python
pa.PAUSE = 3

# Passo 1

pa.hotkey('ctrl', 't')
pc.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga')
pa.hotkey('ctrl', 'v')
pa.press('enter')

time.sleep(10)
# Passo 2
pa.PAUSE = 5
pa.click(x=377, y=272, clicks=2)

time.sleep(10)

# Passo 3
pa.PAUSE = 2
pa.click(x=385, y=340)
pa.click(x=1157, y=152)
pa.click(x=960, y=556)

time.sleep(10)
```

```python
# Passo 4
import pandas as pd

tabela = pd.read_excel(r"C:\Users\Acer\Downloads\Vendas - Dez.xlsx")

display(tabela)

faturamento = tabela['Valor Final'].sum()
quantidade = tabela['Quantidade'].sum()
print(faturamento)
print(quantidade)
```

```python
# Passo 5
pa.PAUSE = 5
time.sleep(10)
pa.hotkey('ctrl', 't')
pc.copy('https://mail.google.com/mail/u/0/#inbox')
pa.hotkey('ctrl', 'v')
pa.press('enter')
pa.PAUSE = 15
time.sleep(15)
pa.click(x=42, y=159)
pa.PAUSE = 5
pa.write('caio201612@gmail.com')
pa.PAUSE = 2
pa.press('tab')
pa.press('tab')
pc.copy('Relatório das compras')
pa.hotkey('ctrl', 'v')
pa.press('tab')

texto = f"""Prezados boa noite!
O faturamento de ontem foi de: {faturamento}
A quantidade de produtos vendidos foi de: {quantidade}

Abraços!
Caio Reis"""

pc.copy(texto)
pa.hotkey('ctrl', 'v')

pa.hotkey('ctrl', 'enter')
```
